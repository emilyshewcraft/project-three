```python
import pandas as pd
from pathlib import Path
```

```python
incident_file = Path("../Datasets/INCIDENT.csv")
incident_df = pd.read_csv(incident_file)
incident_df.head()
```

```python
incident_df.dtypes
```

```python
incident_df['Date'] = pd.to_datetime(incident_df['Date'])

incident_df
```

```python
clean_incident_df=incident_df.drop(['Sources', 'Number_News', 'Reliability', 'Quarter', 'Accomplice', 'Hostages', 'Barricade', 'Officer_Involved', 'Bullied', 'Domestic_Violence', 'Gang_Related', 'Preplanned', 'Shots_Fired', 'Active_Shooter_FBI', 'Location', "Location_Type", 'Time_Period', 'First_Shot', 'Summary', 'Narrative', 'Situation', 'During_School'], axis=1) 
```

```python
clean_incident_df
```

```python
def organize_level(string):
    if string == "Elementary":
        return "Elementary"
    elif string == "High":
        return "High"
    elif string == "Junior High":
        return "Junior High"
    elif string == "K-12":
        return "K-12"
    elif string == "K-8":
        return "K-8"
    elif string == "Middle":
        return "Middle"
    elif string == "Other":
        return "Other"
    else:
        return None

clean_incident_df['School_Level'].apply(organize_level)
clean_incident_df.head()
```

```python
def organize_target(string):
    if string == "Random Shooting":
        return "Random Shooting"
    elif string == "Victims Targeted":
        return "Victims Targeted"
    elif string == "Both":
        return "Both"
    elif string == "Neither":
        return "Neither"
    else:
        return None

clean_incident_df['Targets'].apply(organize_target)
clean_incident_df.dropna(subset=['Targets'], inplace=True)
clean_incident_df.head()
```

```python
clean_incident_df.to_csv("clean_incident.csv", index=False, header=True)
```
