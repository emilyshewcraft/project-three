```python
# Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Save path to the data set as a variable.
shooter_file = Path("../Datasets/SHOOTER.csv")
```

```python
# Use Pandas to the read data.
shooter_df = pd.read_csv(shooter_file)
shooter_df.head()
```

```python
def pull_strings(string):
    if string == "Teen":
        return "Teen"
    elif string == "Adult":
        return "Adult"
    elif string == "Child":
        return "Child"
    elif string == "Minor":
        return "Minor"
    else:
        return None

shooter_df['age_cat'] = shooter_df['age'].apply(pull_strings)
shooter_df.head(10)
```

```python
def replace_strings(string):
    if string == "Teen":
        return 0
    elif string == "Adult":
        return 0
    elif string == "Child":
        return 0
    elif string == "Minor":
        return 0
    elif string == "null":
        return 0
    elif string == "":
        return 0
    else:
        return string

shooter_df['age'] = shooter_df['age'].apply(replace_strings)
shooter_df['age'] = shooter_df['age'].fillna(0)
shooter_df.head(10)
```

```python
shooter_df["age"] = shooter_df["age"].astype(int)
shooter_df.dtypes
```

```python
def categorize(age):
    if age < 12 and age > 0:
        return "Child"
    elif age < 21 and age > 12:
        return "Teen"
    elif age > 21:
        return "Adult"
    else:
        return None

shooter_df['age_cat2'] = shooter_df['age'].apply(categorize)
shooter_df.head(10)
```

```python
shooter_df['age_cat'] = shooter_df['age_cat'].fillna(shooter_df['age_cat2'])
shooter_df.head(10)
```

```python
clean_shooter_df = shooter_df[["incidentid", "schoolaffiliation", "age_cat", "shooterdied"]]
clean_shooter_df.head(10)
```

```python
clean_shooter_df.rename(columns={'incidentid': "Incident_ID"}, inplace=True)
clean_shooter_df.head()
```

```python
clean_shooter_df.to_csv("clean_shooter.csv", index=False, header=True)
```

```python
#Analysis

#Group by whether or not the shooter died

clean_shooter_df.groupby('shooterdied')['Incident_ID'].count().reset_index()
```

```python
#Group by age

clean_shooter_df.groupby('age_cat')['Incident_ID'].count().reset_index()
```

```python
#Group by shooter's affiliation to the school

clean_shooter_df.groupby('schoolaffiliation')['Incident_ID'].count().reset_index()
```
