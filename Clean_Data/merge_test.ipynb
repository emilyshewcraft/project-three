```python
# Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Save path to the data set as a variable.
victim_file = Path("clean_victim.csv")
shooter_file = Path("clean_shooter.csv")
incident_file = Path("clean_incident.csv")
weapon_file = Path("clean_weapon.csv")
```

```python
# Use Pandas to the read data.
victim_df = pd.read_csv(victim_file)
shooter_df = pd.read_csv(shooter_file)
incident_df = pd.read_csv(incident_file)
weapon_df = pd.read_csv(weapon_file)
```

```python
merge_df = pd.merge(incident_df, weapon_df, on="Incident_ID", how="outer")
merge_df
```

```python
merge_df["Incident_ID"].nunique()
```

```python
merge_2_df = pd.merge(merge_df, shooter_df, on="Incident_ID", how="outer")
merge_2_df
```

```python
merge_3_df = pd.merge(merge_2_df, victim_df, on="Incident_ID", how="outer")
merge_3_df
```

```python
merge_3_df.rename(columns={'weapontype': "Weapon_Type",
                                'schoolaffiliation': "Shooter_Cat",
                                'age_cat': "Shooter_Age",
                                'shooterdied': "Shooter_Died",
                                'Injury': "Victim_Injury",
                                'Age': "Victim_Age"}, inplace=True)
merge_3_df
```

```python
merge_3_df["Incident_ID"].nunique()
```

```python
merge_3_df.to_csv("merge_test.csv", index=False, header=True)
```
