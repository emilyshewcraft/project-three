```python
# Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Save path to the data set as a variable.
victim_file = Path("../Original_Datasets/VICTIM.csv")
```

```python
# Use Pandas to the read data.
victim_df = pd.read_csv(victim_file)
victim_df.head()
```

```python
def pull_strings(string):
    if string == "Teen":
        return "Teen"
    elif string == "Adult":
        return "Adult"
    elif string == "Child":
        return "Child"
    elif string == "Officer":
        return "Officer"
    else:
        return ""

victim_df['age_cat'] = victim_df['age'].apply(pull_strings)
victim_df.head(10)
```

```python
def replace_strings(string):
    if string == "Teen":
        return 0
    elif string == "Adult":
        return 0
    elif string == "Child":
        return 0
    elif string == "Officer":
        return 0
    elif string == "null":
        return 0
    elif string == "":
        return 0
    else:
        return string

victim_df['age'] = victim_df['age'].apply(replace_strings)
victim_df.head(10)
```

```python
victim_df['age'] = victim_df['age'].fillna(0)
victim_df["age"] = victim_df["age"].astype(int)
victim_df.dtypes
```

```python
def categorize(age):
    if age < 12 and age > 0:
        return "Child"
    elif age < 21 and age > 12:
        return "Teen"
    elif age > 21:
        return "Adult"
    else:
        return 0

victim_df['age_cat2'] = victim_df['age'].apply(categorize)
victim_df.head(10)
```
