```python
# Dependencies
import pandas as pd
from pathlib import Path
```

```python
shooter_file = Path("../Datasets/SHOOTER.csv")
```

```python
shooter_df = pd.read_csv(shooter_file)
shooter_df.head()
```

```python
shooter_df.loc[shooter_df['age'] == 'Minor', 'age'] = 'Teen'
```

```python
#pull the strings
def pull_strings(string):
    if string == "Teen":
        return "Teen"
    elif string == "Adult":
        return "Adult"
    elif string == "Child":
        return "Child"
    elif string == "null":
        return "null"
    else:
        return None

shooter_df['age_cat'] = shooter_df['age'].apply(pull_strings)
shooter_df['age_cat'].unique()
```

```python
def replace_strings(string):
    if string == "Teen":
        return 0
    elif string == "Adult":
        return 0
    elif string == "Child":
        return 0
    elif string == "Minor":
        return 0
    elif string == "null":
        return 0
    elif string == "":
        return 0
    else:
        return string

shooter_df['age'] = shooter_df['age'].apply(replace_strings)
shooter_df['age_cat'].unique()
```

```python


shooter_df['age'] = shooter_df['age'].fillna(0)
shooter_df["age"] = shooter_df["age"].astype(int)
shooter_df.dtypes

```

```python
def categorize(age):
    if age < 12 and age > 0:
        return "Child"
    elif age < 21 and age > 12:
        return "Teen"
    elif age > 21:
        return "Adult"
    else:
        return None

shooter_df['age_cat2'] = shooter_df['age'].apply(categorize)
shooter_df['age_cat2'].unique()
```

```python


shooter_df['age_cat'] = shooter_df['age_cat'].fillna(shooter_df['age_cat2'])
shooter_df.head(10)

```

```python
#Drop unnecessary columns
clean_shooter_df = shooter_df[['incidentid', 'age_cat', 'schoolaffiliation', 'shooterdied']]
clean_shooter_df.head(20)
```
